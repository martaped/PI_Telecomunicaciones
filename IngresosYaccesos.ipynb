```python
import pandas as pd
df05=pd.read_csv("data/05-Internet_Accesos-por-tecnologia.csv")
df13=pd.read_csv("data/13-Internet_Ingresos.csv")
```

## En este notebook la idea es juntar las dos tablas con informaci칩n por trimestre y por tipo de conectividad (ADSL,Cablemodem, Fibra 칩ptica, Wireless) , a los fines de obtener ingresos discriminados por tipo de conectividad

Realizo un merge de las 2 tablas para evaluar ingresos y accesos

```python
# Realizar el merge por las columnas 'A침o' y 'Trimestre'
df_merged = pd.merge(df05, df13, on=['A침o', 'Trimestre'], how='inner')
```

```python
df_merged.columns
```

Voy a multiplicar los valores por mil de aquellos menores a 10 porque es obvio que hay un salto de valores,
por ejemplo passa de 949 a 1 en vez de pasar a mil


```python
# Multiplicar las columnas por mil si su valor es menor a 10
# Lista de columnas a multiplicar
columnas_a_multiplicar = ['ADSL', 'Cablemodem', 'Fibra 칩ptica', 'Wireless', 'Otros']

# Multiplicar las columnas por mil si su valor es menor a 10
df_merged[columnas_a_multiplicar] = df_merged[columnas_a_multiplicar].applymap(lambda x: x * 1000 if x < 10 else x)
```

```python
del df_merged["Periodo_x"]
```

```python
# Renombrar la columna 'periodo_x' a 'periodo'
df_merged.rename(columns={'Periodo_y': 'Periodo'}, inplace=True)
```

```python
df_merged.to_csv("data/Accesos_ingresos.csv",index=False)
```
