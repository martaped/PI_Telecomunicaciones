## Transformaciones necesarias para mostrar ubicacion geografica de localidades con conectividad por si o por no , discriminada por tipo de conectividad

```python
import pandas as pd
```

Leo la tabla de localidades ya preparada por mi en EDA.py

```python

loc_inter=pd.read_csv("data/16-Localcon_Internet.csv")
```

Leo la tabla de Información Censal del gobierno que contiene latitud y longitud por localidad

```python
loc=pd.read_csv("data/localidades.csv")
```

Renombro columnas partidos y localidades a los fines de normalizar con ambas tablas

```python
loc = loc.rename(columns={'departamento_nombre': 'Partido',})
```

```python
loc = loc.rename(columns={'provincia_nombre': 'Provincia',})
```

```python
loc = loc.rename(columns={'localidad_censal_nombre': 'Localidad',})
```

Borro columnas innecesarias

```python
loc.drop(['categoria',
        'fuente', 'id', 'localidad_censal_id',
        'municipio_id', 
       'provincia_id', 'provincia_nombre'])
```

Como contienen parentesis los nombre se los saco

```python
# Definir una función para quitar el contenido entre paréntesis
def limpiar_localidad(localidad):
    return localidad.split('(')[0].strip()

# Aplicar la función a la columna 'Localidad'
loc_inter['Localidad'] = loc_inter['Localidad'].apply(limpiar_localidad)


# Utilizar str.replace para quitar el contenido entre paréntesis
loc_inter['Localidad'] = loc_inter['Localidad'].str.replace(r'\([^)]*\)', '').str.strip()
```

cambio los nombres de la provincia a formato mayuscula-minuscula

```python
# Convertir cada palabra a formato "primera letra mayúscula y el resto minúscula"
loc_inter['Provincia'] = loc_inter['Provincia'].apply(lambda x: ' '.join(word.capitalize() for word in x.split()))
```

Realizo el merge de loc_inter que es mi tabla y loc que es la tabla con las ubicaiones

```python
# Juntar los DataFrames en base a las columnas 'Partido' y 'Localidad'

df_junto = pd.merge(loc_inter, loc, left_on=['Provincia','Partido', 'Localidad'], right_on=['Provincia','Partido', 'Localidad'], how='left')
```
